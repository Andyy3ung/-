```python
%matplotlib inline
import pandas as pd
import numpy as np
dftrain=pd.read_csv('train.csv')
dftest=pd.read_csv('test.csv')
print('训练集前10行')
print(dftrain.head(10))
print('测试集前10行')
print(dftest.head(10))

deadm=dftrain.groupby(['Survived','Sex']).get_group((0,'male')).count()['Name']
deadf=dftrain.groupby(['Survived','Sex']).get_group((0,'female')).count()['Name']
surm=dftrain.groupby(['Survived','Sex']).get_group((1,'male')).count()['Name']
surf=dftrain.groupby(['Survived','Sex']).get_group((1,'female')).count()['Name']

df0 = pd.DataFrame({'Dead':[deadm,deadf]},index=['Male','female'])
df0.plot.pie(subplots=True,autopct='%1.1f%%')

df1 = pd.DataFrame({'Survived':[surm,surf]},index=['Male','Female'])
df1.plot.pie(subplots=True,autopct='%1.1f%%')

dftrain['AgeGroup']=pd.cut(dftrain['Age'],5)
df=dftrain.groupby(['Sex'])
dfm=df.get_group('male')
dff=df.get_group('female')
dfma=dfm.groupby('AgeGroup')
dffa=dff.groupby('AgeGroup')


rsm=dfma['Survived'].agg(np.mean)
rsf=dffa['Survived'].agg(np.mean)
rs1=pd.DataFrame(rsm)
rs2=pd.DataFrame(rsf)
rs=pd.merge(rs1,rs2,on=['AgeGroup'],how='outer')
rs.columns = ['Survived Male', 'Survived Female']


rs.plot.bar()
```
